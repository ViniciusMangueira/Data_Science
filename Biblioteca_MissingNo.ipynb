```
#Importar as bibliotecas 
import pandas as pd
import numpy as np
import missingno as msno # Biblioteca MissingNo
%matplotlib inline
```

```
# Importar o arquivo
df = pd.read_csv("/content/netflix_titles.csv")
```

```
# Ver as 5 primeiras entradas
df.head(5)
```

```
#Ver a quantidade de valores faltantes de cada atributo
df.isnull().sum().sort_values(ascending=False)
```

```
# Matriz dos dados faltantes
msno.matrix(df, figsize=(10,8));
```

```
# Gráficos dos missing values:
msno.bar(df, figsize=(10,8));
```

```
#Criando novo dataset
df2 = df.copy()  #Cópia do dataset original 
df2.dropna(subset=['director'], inplace=True) # Excluindo os dados faltantes do atributo director do dataset 'df2'
df2.isnull().sum().sort_values(ascending=False) #Verificando a quantidade de dados faltantes, agora, do df2
```

```
print('Formato do dataset `df` : ',df.shape)
print('Formato do dataset `df2`: ',df2.shape)
```

```
# Matriz dos dados faltantes
msno.matrix(df2, figsize=(10,8));
```

```
# Gráficos dos missing values:
msno.bar(df2, figsize=(10,8));
```

```
#Mapa de calor
msno.heatmap(df, figsize=(10,8));
```

```
#Dendrograma
msno.dendrogram(df, figsize=(10,8));
```
